```python
import scipy.io as sio
import pandas as pd
import numpy as np
import glob2 as glob
from pathlib import Path
```

# Convert all .mat subject data files into one single panda dataframe.

```python
pd.set_option('display.max_columns', 300)
pd.set_option('display.max_rows', None)
```

```python
p = "taskReversalLearn/data/"
allFiles = []
```

```python
# import all .mat files into the directory
for path in Path(p).rglob('*.mat'):
    allFiles.append(str(path))
```

```python
print (allFiles[10])
```

```python
subject = []
choice = []
outcome = [] 
RT = []
env = []
uncertainty =[] 
feedback = []
```

```python
for f in allFiles:
    mat = sio.loadmat(f, squeeze_me=True)
    m = mat['data']
    choice.extend(m.tolist()[1])
    outcome.extend(m.tolist()[2])
    RT.extend(m.tolist()[3])
    subject.extend([m.tolist()[0]]*135)
    deets = m.tolist()[7]
    env.extend(([deets.tolist()[1]])*135) #volatility type
    uncertainty.extend(([deets.tolist()[2]])*135) #0.7 and 0.3
    feedback.extend(deets.tolist()[5])
```

```python
df = pd.DataFrame(list(zip(subject, choice, outcome, RT, env, uncertainty, feedback)), 
               columns =['SubjectID', 'choice', "outcome", "RT", "env", "uncertainty", "feedback"]) 
```
